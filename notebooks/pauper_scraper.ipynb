```python
from scraper import urls, event_processor

page_url = urls.build_decklists_url(
    year=2026,
    month=6,
    format="Pauper",
    event_type="Challenge"
)


all_urls = urls.get_page_event_urls(page_url)


pauper_urls = urls.filter_events(
    all_urls,
    "pauper-challenge"
)
```

```python
len(pauper_urls)
```

```python
all_events = []

for url in pauper_urls:
    try:
        result = event_processor.scrape_event(url)
        all_events.append(result)

        print("processed:", result["event"]["event_id"])

    except Exception as e:
        print("failed:", url, e)
```

```python
import pandas as pd

events = pd.DataFrame([e["event"] for e in all_events])

standings = pd.DataFrame(
    [row for e in all_events for row in e["standings"]]
)

decklists = pd.DataFrame(
    [row for e in all_events for row in e["decklists"]]
)

brackets = pd.DataFrame(
    [row for e in all_events for row in e["brackets"]]
)
```

```python
len(all_events)
```

```python
events.to_csv("../data/processed/events.csv", index=False)
standings.to_csv("../data/processed/standings.csv", index=False)
decklists.to_csv("../data/processed/decklists.csv", index=False)
brackets.to_csv("../data/processed/brackets.csv", index=False)
```
