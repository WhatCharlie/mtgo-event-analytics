```python
from scraper import urls, event_processor, file_io

page_url = urls.build_decklists_url(
    year=2026,
    month=6,
    format="Pauper",
    event_type="Challenge"
)


all_urls = urls.get_page_event_urls(page_url)


pauper_urls = urls.filter_events(
    all_urls,
    "pauper-challenge"
)
```

```python
import scraper.file_io

print(scraper.file_io.__file__)
```

```python
len(pauper_urls)
```

```python
all_events = []

for url in pauper_urls:
    try:
        result = event_processor.scrape_event(url)
        all_events.append(result)

        print("processed:", result["event"]["event_id"])

    except Exception as e:
        print("failed:", url, e)
```

```python
import pandas as pd

events = pd.DataFrame([e["event"] for e in all_events])

standings = pd.DataFrame(
    [row for e in all_events for row in e["standings"]]
)

decklists = pd.DataFrame(
    [row for e in all_events for row in e["decklists"]]
)

brackets = pd.DataFrame(
    [row for e in all_events for row in e["brackets"]]
)
```

```python
len(all_events)
```

```python
events = file_io.update_csv(
    events,
    "../data/processed/events.csv",
    "event_id"
)

standings = file_io.update_csv(
    standings,
    "../data/processed/standings.csv",
    ["event_id", "player"]
)

decklists = file_io.update_csv(
    decklists,
    "../data/processed/decklists.csv",
    ["event_id", "player"]
)

brackets = file_io.update_csv(
    brackets,
    "../data/processed/brackets.csv",
    ["event_id", "round", "player1", "player2"]
)
```

```python
print(len(events))
print(len(standings))
print(len(decklists))
print(len(brackets))
```
